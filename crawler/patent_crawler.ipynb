```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

# Set up the WebDriver
driver_path = '/Users/munseunghyeon/Downloads/chromedriver-mac-arm64/chromedriver'  # 최신 ChromeDriver 경로로 업데이트
service = Service(driver_path)
driver = webdriver.Chrome(service=service)

def get_patent_len():
    # Locate the tbody element
    tbody = driver.find_element(By.XPATH, '//*[@id="tableData"]/tbody')
    # Find all tr elements within the tbody
    rows = tbody.find_elements(By.TAG_NAME, "tr")
    print("Number of rows:", len(rows))
    return len(rows)

try:
    # Open the target website
    driver.get("https://iprsearch.ipindia.gov.in/publicsearch")

    # Wait for the search field to be present
    search_field = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//*[@id="TextField1"]'))
    )

    # Enter a search term
    search_field.send_keys("pet")

    # Wait for the CAPTCHA input to be present
    captcha_input = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//*[@id="CaptchaText"]'))
    )

    # Inform the user to enter the CAPTCHA manually
    print("Please enter the CAPTCHA in the browser and press Enter here to continue...")
    captcha_input.send_keys(input())

    # After manual CAPTCHA entry, submit the form
    search_button = driver.find_element(By.XPATH, '//*[@id="home"]/div[18]/div[2]/div/div/div[1]/div[3]/input')
    search_button.click()

    # Wait for the total_patent element to be present
    total_patent = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//*[@id="header"]/div[4]/div/div[1]/div[2]'))
    )

    # Extract all patent information
    patent_contents = []
    titles = []
    app_nums = []
    dates = []

    # Get the total number of pages
    last_page = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//*[@id="tableData"]/tfoot/tr/td/table/tbody/tr/th[2]/button[4]'))
    )
    last_page.click()

    len_page = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//*[@id="tableData"]/tfoot/tr/td/table/tbody/tr/th[2]/span'))
    )

    total_pages = int(len_page.text.split()[-1])

    # Iterate through each page
    for page in range(total_pages):
        # Iterate through each patent in the current page
        for num_p in range(get_patent_len()):
            # Switch back to main page
            driver.switch_to.window(driver.window_handles[0])

            # Locate and click the patent link
            patent = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, f'//*[@id="tableData"]/tbody/tr[{num_p + 1}]/td[1]/button'))
            )
            patent.click()

            # Switch to the new tab
            WebDriverWait(driver, 10).until(EC.number_of_windows_to_be(2))
            driver.switch_to.window(driver.window_handles[1])

            # Extract patent details
            try:
                patent_content = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.XPATH, '//*[@id="COMPLETE_SPECIFICATION"]'))
                )
            except:
                patent_content = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.XPATH, '//*[@id="home"]/table/tbody/tr[16]/td'))
                )

            title = driver.find_element(By.XPATH, '//*[@id="home"]/table/tbody/tr[1]/td[2]').text
            app_num = driver.find_element(By.XPATH, '//*[@id="home"]/table/tbody/tr[5]/td[2]').text
            date = driver.find_element(By.XPATH, '//*[@id="home"]/table/tbody/tr[3]/td[2]').text

            patent_contents.append(patent_content.text)
            titles.append(title)
            app_nums.append(app_num)
            dates.append(date)

            time.sleep(1)

            # Close the tab
            driver.close()

        # Switch back to main page
        driver.switch_to.window(driver.window_handles[0])

        # Click the "Previous" button to go to the previous page
        if page < total_pages - 1:
            previous_button = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, '//*[@id="tableData"]/tfoot/tr/td/table/tbody/tr/th[2]/button[2]'))
            )
            previous_button.click()
            time.sleep(4)

    # Create a DataFrame and save to CSV
    df = pd.DataFrame({
        "Title": titles,
        "Application Number": app_nums,
        "Date": dates,
        "Patent Content": patent_contents
    })
    df.to_csv("patents.csv", index=False)

finally:
    # Close the WebDriver
    driver.quit()
    print("done")
```
